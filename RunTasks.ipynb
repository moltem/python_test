```python
import os

from Json2Html import Json2Html

 
########################################

#####################TASK1##############

########################################

t1_json_file_path_in = "Task_1_in.json"

t1_json_file_path_out = "Task_1_out.html"

str_Task_1_in = """

                    [

                        {

                            "title":"Title #1",

                            "body":"Hello world 1!"    

                        },{

                            "title":"Title #2",

                            "body":"Hello world 2!"      

                        }

                    ]

                """

str_Task_1_in_33 = """
                        {

                            "title":"Title #1",

                            "body":"Hello world 1!"    

                        }
                """

str_Task_1_out = """<h1>Title #1</h1><p>Hello world 1!</p><h1>Title #2</h1><p>Hello world 2!</p>"""

 

if os.path.exists(t1_json_file_path_in):

    os.remove(t1_json_file_path_in)

   

if os.path.exists(t1_json_file_path_out):

    os.remove(t1_json_file_path_out)

 

with open(t1_json_file_path_in,'w') as file:

    file.write(str_Task_1_in)

    

with open(t1_json_file_path_out,'w') as file:

    file.write(str_Task_1_out)   

 

html_generator = Json2Html()

print("+++++++++++++++++task_1_json2html+++++++++++++++++++++++++++++++++++++")

print(html_generator.task_1_json2html(t1_json_file_path_in))

print("++++++++++++++++++task_1_json2html++++++++++++++++++++++++++++++++++++")

 

if str_Task_1_out != html_generator.task_1_json2html(t1_json_file_path_in):

    raise ValueError('Results are different!')

 

########################################

#####################TASK2##############

########################################

t2_json_file_path_in = "Task_2_in.json"

t2_json_file_path_out = "Task_2_out.html"

str_task_2_in = """

                [

                    {

                        "h3": "Title 1",

                        "div" : "Hello, World1!"

                    },

                    {

                        "h3": "Title 2",

                        "div" : "Hello, World2!"

                    }

                ]

            """

str_task_2_out = """<h3>"Title 1"</h3><div>"Hello, World1!"</div><h3>"Title 2"</h3><div>"Hello, World2!"</div>"""

 

if os.path.exists(t2_json_file_path_in):

    os.remove(t2_json_file_path_in)

   

if os.path.exists(t2_json_file_path_out):

    os.remove(t2_json_file_path_out)

 

   

with open(t2_json_file_path_in,"w") as file:

    file.write(str_task_2_in)

 

with open(t2_json_file_path_out,"w") as file:

    file.write(str_task_2_out)

 

html_generator_t2 = Json2Html()

print("++++++++++++++++++++++task_2_json2html++++++++++++++++++++++++++++++++")

print(html_generator_t2.task_2_json2html(t2_json_file_path_in))

print("++++++++++++++++++++++task_2_json2html++++++++++++++++++++++++++++++++")

 

############################################

#####################TASK3##############

############################################

t3_json_file_path_in = "Task_3_in.json"

t3_json_file_path_out = "Task_3_out.html"

str_task_3_in = """

                [

                    {

                        "h3": "Title 1",

                        "div" : "Hello, World1!"

                    },

                    {

                    "h3": "Title 2",

                    "div" : "Hello, World2!"

                    }

                ]

            """

 

str_task_3_out = """<ul><li><h3>"Title 1"</h3><div>"Hello, World1!"</div></li><li><h3>"Title 2"</h3><div>"Hello, World2!"</div></li></ul>"""

 

if os.path.exists(t3_json_file_path_in):

    os.remove(t3_json_file_path_in)

   

if os.path.exists(t3_json_file_path_out):

    os.remove(t3_json_file_path_out)

 

 

with open(t3_json_file_path_in,"w") as file:

    file.write(str_task_3_in)

 

with open(t3_json_file_path_out,"w") as file:

    file.write(str_task_3_out)

   

html_generator_t3 = Json2Html()

print("++++++++++++++++++++++++++task_3_4_json2html++++++++++++++++++++++++++++")

print(html_generator_t3.task_3_4_json2html(t3_json_file_path_in))

print("+++++++++++++++++++++++++++task_3_4_json2html+++++++++++++++++++++++++++")

 

############################################

#####################TASK4##############

############################################

t4_json_file_path_in = "Task_4_in.json"

t4_json_file_path_out = "Task_4_out.html"

str_task_4_in = """

                [

                    {

                        "span": "Title 1",

                        "content" : [{

                                        "p":"exampl 1",

                                        "header":"header1"

                                    }]

                    },

                    {"div" : "div 1"}

                ]

            """

str_task_4_out = """<ul><li><span>"Title 1"</span><content><ul><li><p>"exampl 1"</p><header>header1</header></li></ul></content></li><li><div>"div 1"</div></li></ul>"""

 

if os.path.exists(t4_json_file_path_in):

    os.remove(t4_json_file_path_in)

   

if os.path.exists(t4_json_file_path_out):

    os.remove(t4_json_file_path_out)

 

 

with open(t4_json_file_path_in,"w") as file:

    file.write(str_task_4_in)

 

with open(t4_json_file_path_out,"w") as file:

    file.write(str_task_4_out)

 

html_generator_t4 = Json2Html()

print("++++++++++++++++++++++++++task_3_4_json2html++++++++++++++++++++++++++++++")

print(html_generator_t4.task_3_4_json2html(t4_json_file_path_in))

print("+++++++++++++++++++++++++++task_3_4_json2html+++++++++++++++++++++++++++++")

 

############################################

#####################TASK5##############

############################################

t5_json_file_path_in = "Task_5_in.json"

t5_json_file_path_out = "Task_5_out.html"

str_task_5_in = """{
                        "p.my-class#my-id": "hello",

                        "p.my-class.my-class2":"example<a>asd</a>"
                     }

            """

str_task_5_out = """<p id="my-id" class="my-class">hello</p><p class="my-class my-class2">example<a>asd</a></p>"""

 

if os.path.exists(t5_json_file_path_in):

    os.remove(t5_json_file_path_in)

   

if os.path.exists(t5_json_file_path_out):

    os.remove(t5_json_file_path_out)

 

with open(t5_json_file_path_in,"w") as file:

    file.write(str_task_5_in)

 

with open(t5_json_file_path_out,"w") as file:

    file.write(str_task_5_out)

 

html_generator_t5 = Json2Html()

print("++++++++++++++++++++++task_5_json2html++++++++++++++++++++++++++++++++")

print(html_generator_t5.task_5_json2html(t5_json_file_path_in))

print("++++++++++++++++++++++++task_5_json2html++++++++++++++++++++++++++++++")
```
